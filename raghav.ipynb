# C: Conditional Probability

**Q8.** In a class of 40 students, 25 passed math, 18 passed science, and 10 passed both. Find the probability that a student passed science given that they passed math.

```python
# do not run this cell
```

```python
# Write your solution here
P_Math=25/40
P_Math_And_Science=10/40
P_Science_Given_Math=P_Math_And_Science/P_Math
P_Science_Given_Math
```

```python
students = list(range(1, 41))
M = set(range(1, 26))
both = set(range(1, 11))
science_only = set(range(26, 34))

S = both | science_only


N7 = 1_000_000
count_math = 0
count_sci_and_math = 0

for _ in range(N7):
    student1 = np.random.choice(students)
    if student1 in M:
        count_math += 1
        if student1 in S:
            count_sci_and_math += 1

res7 = count_sci_and_math / count_math
print(f"Exact Result: {P_Science_Given_Math}\nUsing Simulation (N=1000000): {res7}")
```
